```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
products = np.array(["Product A", "Product B", "Product C", "Product D"])
sales_amount = np.array([400, 300, 500, 600])
```

```python
plt.bar(products, np.sort(sales_amount))
plt.title('Sales of products')
plt.ylabel('Sales amount')
plt.ylabel('Products')
plt.grid()
```

We can see that the product D has highest sales and product A has lowest sales.
