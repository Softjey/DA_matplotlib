```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = np.array(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]);
visiters_amount = np.array([120, 150, 130, 170, 200, 230, 180])
```

```python
plt.plot(days, visiters_amount, marker="o")
plt.title('Unique visiters by week')
plt.xlabel('Days of week')
plt.ylabel('Visiters amount')
plt.yticks(np.arange(visiters_amount.min() - 40, visiters_amount.max() + 40, 20))
plt.grid()
```

We can see that the maximum of unique visiters amount on Saturday, and the minimum on Monday. Also visiters amount is increasing from Monday to Saturday.
